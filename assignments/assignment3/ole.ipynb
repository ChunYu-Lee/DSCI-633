```python
import pandas as pd
import numpy as np
from collections import Counter

data_train = pd.read_csv("../data/audiology_train",header=None)
# Separate independent variables and dependent variables
independent = range(69)
X = data_train[independent]
y = data_train[70]


# turn nan in y into N as a type of category
if y.isnull().values.any() == True:
      y = y.fillna("N")

        
# turn nan into abc as a type of category
if X.isnull().values.any() == True:
    X = X.fillna("abc")
# print(X.isnull().values.any())
# print(X.isnull().sum())


# num = {}
# for key in X:
#     num[key] = list(set(X[key]))

# print(num)
# for key in X:
#     for i in num[key]:
#         print(i)

classes_ = list(set(list(y)))
# for calculation of P(y)
P_y = Counter(y)
# self.P[yj][Xi][xi] = P(xi|yi) where Xi is the feature name and xi is the feature value, yj is a specific class label
P = {}

# get the all the y index
target_idx = {}

for i in classes_:
    target_list_idx = []
    for j in range(len(y)):
        if y[j] == i:
            target_list_idx.append(j)
    target_idx[i] = target_list_idx
                          
#pre-calculate
for label in classes_:
    independent = {}
    key_dict = {}
    
    for key in X:
        num = list(set(X[key]))
        Xi_value = Counter(X[key][target_idx[label]])
        for i in range(len(num)):
            if (num[i] not in Xi_value):
                Xi_value[num[i]] = 0
        independent[key] = Xi_value

#         for i in num[key]:
#             if (i not in Xi_value):
#                 Xi_value[i] = 0
#         independent[key] = Xi_value
        print(independent)
        prob = {}
        for i in independent[key]:
            lap = (independent[key][i]+ 1)/ (P_y[label] + (1 * len(independent[key])) )
            prob[i] = lap
        key_dict[key] = prob
    P[label] = key_dict
            
P = pd.DataFrame(P)
            
# print(P)
# print(X[3][target_idx["cochlear_age"]])
```
